```python
#step1: collecting the data
import pandas as pd
df = pd.read_csv("ToyotaCorolla - MLR.csv")

df.shape
df.head()
```

```python
df.info()
```

```python

#step2: EDA --> boxplot, histogrmam, scatter plot, 

#shown in mlr models file
```

```python
#step3: Data cleaning

df.info()

##need to observe if fany null values 
##since there are not any null values present we are continuing with the data presented
```

```python
##CONTINUOUS DATA COLUMNS
df_cont=df[['Age_08_04','KM','HP','Automatic','cc','Doors','Cylinders','Gears','Weight']]
```

```python
##STEP 4
#STEP 4 a DATA PRE-PROCESSING
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

df_cont['Age_08_04'] = SS.fit_transform(df_cont[['Age_08_04']])
df_cont['KM'] = SS.fit_transform(df_cont[['KM']])
df_cont['HP'] = SS.fit_transform(df_cont[['HP']])

df_cont['Automatic'] = SS.fit_transform(df_cont[['Automatic']])
df_cont['cc'] = SS.fit_transform(df_cont[['cc']])
df_cont['Doors'] = SS.fit_transform(df_cont[['Doors']])

df_cont['Cylinders'] = SS.fit_transform(df_cont[['Cylinders']])
df_cont['Gears'] = SS.fit_transform(df_cont[['Gears']])
df_cont['Weight'] = SS.fit_transform(df_cont[['Weight']])



df_cont.head()
```

```python

##step 4 B
##split X and Y
##performing label encoding on Categorical data
## STEP 4 B Fuel_Type is label encoding column
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df["Fuel Type Encoded"]=LE.fit_transform(df['Fuel_Type'])
df["Fuel Type Encoded"]
```

```python
###combining categorical and continuous columns in order to get preprocesing of X variables
#STEP 4 C 
##converting the above array into DataFrame
df_cat = pd.DataFrame(df['Fuel Type Encoded'], columns=['Fuel Type Encoded'])
df_cat
```

```python
##step 4 D
##FINALLY we are combining all the pre-processed variables
##those are X continuous variables and Categorical variables
###we consider the new data that is df_newX for training and testing X variables
df_newX=pd.concat([df_cont,df_cat],axis=1)
df_newX
```

```python
##step 5 
Y=df["Price"]
Y
```

```python

#step 6
#Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df_newX,Y,test_size=0.25)
```

```python

#step
# STEP 7: model fitting
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test  = model.predict(X_test)
```

```python
#STEP 8 
# data validation
from sklearn.metrics import mean_squared_error
train_rmse = mean_squared_error(Y_train,Y_pred_train,squared=False)
print("Training Error:",train_rmse)
test_rmse = mean_squared_error(Y_test,Y_pred_test,squared=False)
print("Test Error:",test_rmse)
```

```python
#STEP 9
#=========================================================
# cross validation
#=========================================================

training_err = []
test_err = []

for i in range(1,101):
    X_train,X_test,Y_train,Y_test = train_test_split(df_newX,Y, test_size=0.25,random_state=i)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test  = model.predict(X_test)
    training_err.append(mean_squared_error(Y_train,Y_pred_train,squared=False))
    test_err.append(mean_squared_error(Y_test,Y_pred_test,squared=False))

import numpy as np
print("Cross validation - Training error:" ,np.mean(training_err).round(2))
print("Cross validation - Test error:" ,np.mean(test_err).round(2))

#=====================================================================
```

```python
LR_coef = model.coef_
LR_coef
#=====================================================================
# LASSO REGRESSION
from sklearn.linear_model import Lasso
X_train,X_test,Y_train,Y_test = train_test_split(df_newX,Y,test_size=0.25,random_state=100)

model = Lasso(alpha=1)

model.fit(X_train,Y_train)

model.intercept_
coeff = pd.concat([pd.DataFrame(list(df_newX)),pd.DataFrame(LR_coef),pd.DataFrame(model.coef_)],axis=1)
coeff

#=====================================================================
```
